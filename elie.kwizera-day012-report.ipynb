### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

```python
# Exercise 10
```

---

````python
x = np.linspace(-10, 10, 400)
y1 = x
y2 = x**2
y3 = x**3
plt.figure(figsize=(8,6))
plt.plot(x, y1, color='red', linestyle='solid', label='y = x')
plt.plot(x, y2, color='green', linestyle='--', label='y = x^2')
plt.plot(x, y3, color='blue', linestyle=':', label='y = x^3')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x, y = x^2, and y = x^3')
plt.legend()
````

### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.

```python
# Exercise 16
```

---

````python
categories = ['X', 'Y', 'Z']
group_a = [3, 5, 7]
group_b = [2, 6, 4]
plt.figure(figsize=(8, 6))
plt.bar(categories, group_a, label='Group A', color='lightblue')
plt.bar(categories, group_b, bottom=group_a, label='Group B', color='lightgreen')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart of Group A and Group B')
plt.legend()
plt.show()
````

### Exercise 20:
Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.

```python
# Exercise 20
```

---

````python
x = np.linspace(0, 2 * np.pi, 100)
sin_wave = np.sin(x)
cos_wave = np.cos(x)
plt.figure(figsize=(8, 6))
plt.plot(x, sin_wave, linestyle='dashed', color='green', label='Sine Wave')
plt.plot(x, cos_wave, linestyle='-', color='magenta', label='Cosine Wave')
plt.title('Sine and Cosine Waves')
plt.legend()
````

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.

```python
# Exercise 21
```

---

````python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contour)
plt.title('Contour Plot of z = sin(x) + cos(y)')
````
